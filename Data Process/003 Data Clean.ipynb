```python
import pandas as pd
```

```python
file_path = "~/Desktop/Spring 2024/POLI179/poli 179 data/People's daily/filtered_data_2004_2017.csv"
filtered_data = pd.read_csv(file_path)
```

```python
train_df=pd.read_csv("~/Desktop/Spring 2024/POLI179/poli 179 data/People's daily/Train/all_train_df.csv")
```

```python
import re
def replace_punctuation(example):
    example = list(example)
    pre = ''
    cur = ''
    for i in range(len(example)):
        if i == 0:
            pre = example[i]
            continue
        pre = example[i-1]
        cur = example[i]
        if re.match("[\u4e00-\u9fa5]", pre):
            if re.match("[\u4e00-\u9fa5]", cur):
                continue
            elif cur == ',':
                example[i] = '，'
            elif cur == '.':
                example[i] = '。'
            elif cur == '?':
                example[i] = '？'
            elif cur == ':':
                example[i] = '：'
            elif cur == ';':
                example[i] = '；'
            elif cur == '!':
                example[i] = '！'
            elif cur == '"':
                example[i] = '”'
            elif cur == "'":
                example[i] = "’"
    return ''.join(example)
```

```python
# Applying the replacement function to the 'TextBody' column
rep_filtered_data = filtered_data[['TextBody', 'Date', 'Year','id']].copy()
rep_filtered_data['TextBody'] = rep_filtered_data['TextBody'].apply(replace_punctuation)
```

```python
print(rep_filtered_data['TextBody'].iloc[1])
```

```python
rep_train_df = train_df[['TextBody', 'id']].copy() 
rep_train_df['TextBody'] = rep_train_df['TextBody'].apply(replace_punctuation)
```

```python
print(rep_train_df['TextBody'].iloc[1])
```

```python
def clean_text(text):
    """
    Clean the input text by performing several common preprocessing steps tailored for Chinese text.
    
    Parameters:
        text (str): The original Chinese text data.
        
    Returns:
        str: The cleaned Chinese text data.
    """
    # Remove Latin script dates (if present in your data and not relevant)
    text = re.sub(r"\d{4}-\d{2}-\d{2}", "", text)
    # Remove or replace non-Chinese characters and keep common punctuation
    text = re.sub(r"[^\u4e00-\u9fff，。；！？：、“”《》（）【】]", "", text)
    # Remove excessive spaces (if any spaces are present; might be less relevant in Chinese text)
    text = re.sub(r"\s+", " ", text).strip()
    
    return text
```

```python
clean_filtered_data = rep_filtered_data[['TextBody', 'Date', 'Year','id']].copy()
clean_filtered_data['TextBody'] = rep_filtered_data['TextBody'].apply(clean_text)
```

```python
print(clean_filtered_data['TextBody'].iloc[1])
```

```python
clean_train_df = rep_train_df[['id', 'TextBody']].copy()
clean_train_df['TextBody'] = rep_train_df['TextBody'].apply(clean_text)
```

```python
print(clean_train_df['TextBody'].iloc[1])
```

```python
clean_filtered_data.to_csv("~/Desktop/Spring 2024/POLI179/poli 179 data/People's daily/final_data.csv", index=False)
```
